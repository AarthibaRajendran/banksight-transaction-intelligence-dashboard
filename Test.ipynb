```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS accounts (
    customer_id VARCHAR(10),
    account_balance DECIMAL(12,2),
    last_updated DATETIME,
               
    FOREIGN KEY (customer_id)
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS transactions (
    txn_id VARCHAR(10) PRIMARY KEY,
    customer_id VARCHAR(10),
    txn_type VARCHAR(50),
    amount DECIMAL(12,2),
    txn_time DATETIME,
    status VARCHAR(50),
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS loans (
    loan_id INT PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id INT,
    branch VARCHAR(50),
    loan_type VARCHAR(50),
    loan_amount INT,
    interest_rate DECIMAL(12,2),
    loan_term_months INT,
    start_date DATE,
    end_date DATE,
    loan_status VARCHAR(50)
               
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS credit_cards (
    card_id INT PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id INT,
    branch VARCHAR(50),
    card_number VARCHAR(20),
    card_type VARCHAR(50),
    card_network VARCHAR(50),
    credit_limit INT,
    current_balance DECIMAL(12,2),
    issued_date DATE,
    expiry_date DATE,
    status VARCHAR(50),
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers( customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS branches (
    branch_id INT PRIMARY KEY,
    branch_name VARCHAR(50),
    city VARCHAR(50),
    manager_name VARCHAR(50),
    total_employees int,
    branch_revenue DECIMAL(12,2),
    opening_date DATE,
    performance_rating int
)
""")
```

```python

cursor.execute("""
CREATE TABLE IF NOT EXISTS support_tickets (
    ticket_id VARCHAR(10) PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id VARCHAR(10),
    loan_id VARCHAR(50) NULL,
    branch_name VARCHAR(50),
    issue_category VARCHAR(50),
    description  TEXT,
    date_opened DATE,
    date_closed DATE,
    priority VARCHAR(50),
    status VARCHAR(50),
    resolution_remarks TEXT,
    support_agent VARCHAR(50),
    channel VARCHAR(50),
    customer_rating int,
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

##### 1 CUSTOMER & ACCOUNT ANALYSIS

Q1: How many customers exist per city, and what is their average account balance?

```python

cust_acc = pd.merge(customers, accounts, on='customer_id', how='left')

result = cust_acc.groupby('city').agg(
    total_customers=('customer_id', 'count'),
    avg_account_balance=('account_balance', 'mean')
).reset_index()

print(result)
```

Q2: Which account type (Savings, Current, Loan, etc.) holds the highest total balance?

```python
cust_acc = pd.merge(customers, accounts, on='customer_id', how='left')

balance_by_type = cust_acc.groupby('account_type').agg(
    total_balance=('account_balance', 'sum')
).reset_index()

highest_balance_type = balance_by_type.loc[balance_by_type['total_balance'].idxmax()]
print(highest_balance_type)

```

Q3: Who are the top 10 customers by total account balance across all account types?

```python
cust_acc = pd.merge(customers, accounts, on='customer_id', how='left')

total_balance_per_customer = cust_acc.groupby(['customer_id', 'name'])['account_balance'].sum().reset_index()

top10_customers = total_balance_per_customer.sort_values(by='account_balance', ascending=False).head(10)
print(top10_customers)
```

Q4: Which customers opened accounts in 2023 with a balance above 1,00,000?

```python
cust_acc = pd.merge(customers, accounts, on='customer_id', how='left')

# cust_acc['join_date'] object to Datetime convert

cust_acc['join_date'] = pd.to_datetime(cust_acc['join_date'], errors='coerce')

accounts_2023_high = cust_acc[
    (cust_acc['join_date'].dt.year == 2023) &
    (cust_acc['account_balance'] > 100000)
]

result = accounts_2023_high[['customer_id', 'name', 'account_balance', 'join_date']]
print(result)
```

##### 2 TRANSACTION BEHAVIOR

Q5: What is the total transaction volume (sum of amounts) by transaction type?

```python

txn_summary = transactions.groupby('txn_type').agg(
    total_amount=('amount', 'sum')
).reset_index()

print(txn_summary)
```

Q6: How many failed transactions occurred for each transaction type?

```python

failed_count = transactions[transactions['status'].str.lower() == 'failed'] \
               .groupby('txn_type') \
               .size() \
               .reset_index(name='failed_count')

print(failed_count)
```

Q7: What is the total number of transactions per transaction type?

```python
txn_summary = transactions.groupby('txn_type').agg(
    total_count=('txn_id', 'count')
).reset_index()

print(txn_summary)
```

Q8: Which accounts have 5 or more high-value transactions above 20,000?

```python
# Step 1: Filter high-value transactions (>20,000)
high_txn = transactions[transactions['amount'] > 20000]

# Step 2: Group by account (customer_id) and calculate count & total amount
high_txn_summary = high_txn.groupby('customer_id').agg(
    txn_count=('txn_id', 'count'),
    total_amount=('amount', 'sum')
).reset_index()

# Step 3: Filter accounts with 5 or more high-value transactions
high_value_accounts = high_txn_summary[high_txn_summary['txn_count'] >= 5]

# Step 4: Display result
print(high_value_accounts)
```

##### 3 LOAN INSIGHTS

Q9: What is the average loan amount and interest rate by loan type (Personal, Auto, Home, etc.)?

```python
loan_summary = loans.groupby('loan_type').agg(
    avg_loan_amount=('loan_amount', 'mean'),
    avg_interest_rate=('interest_rate', 'mean')
).reset_index()

print(loan_summary)
```

Q10: Which customers currently hold more than one active or approved loan?

```python

# Step 1: Filter only active or approved loans
active_loans = loans[loans['loan_status'].isin(['Active', 'Approved'])]

# Step 2: Group by customer_id and count number of loans
loan_counts = active_loans.groupby('customer_id').agg(
    num_loans=('loan_id', 'count')
).reset_index()

# Step 3: Filter customers with more than 1 loan
customers_multiple_loans = loan_counts[loan_counts['num_loans'] > 1]

# Step 4: Merge with customer names (optional, to see names)
result = pd.merge(customers_multiple_loans, customers[['customer_id', 'name']], on='customer_id', how='left')

print(result)
```

Q11: Who are the top 5 customers with the highest outstanding (non-closed) loan amounts?

```python
# Step 1: Filter only non-closed loans
non_closed_loans = loans[loans['loan_status'] != 'Closed']

# Step 2: Group by customer_id and sum loan_amount
outstanding = non_closed_loans.groupby('customer_id').agg(
    total_outstanding=('loan_amount', 'sum')
).reset_index()

# Step 3: Sort by total outstanding amount descending
top5_customers = outstanding.sort_values(by='total_outstanding', ascending=False).head(5)

# Step 4: Merge with customer names (optional)
top5_customers = pd.merge(top5_customers, customers[['customer_id', 'name']], on='customer_id', how='left')

print(top5_customers)
```

##### 4 BRANCH & PERFORMANCE

Q12: What is the average loan amount per branch?

Q13: How many customers exist in each age group (e.g., 1825, 2635, etc.)?

```python
# Step 1: Define age bins and labels
age_bins = [0, 18, 25, 35, 45, 60, 120]
age_labels = ['<18', '18-25', '26-35', '36-45', '46-60', '60+']

# Step 2: Create age_group column
customers['age_group'] = pd.cut(
    customers['age'],
    bins=age_bins,
    labels=age_labels,
    right=True
)

# Step 3: Count customers per age group
age_group_count = customers.groupby('age_group').agg(
    total_customers=('customer_id', 'count')
).reset_index()

# Step 4: Display result
print(age_group_count)
```

##### 5 SUPPORT TICKETS & CUSTOMER EXPERIENCE

Q14: Which issue categories have the longest average resolution time?

```python
# Ensure date columns are datetime
support_tickets['date_opened'] = pd.to_datetime(support_tickets['date_opened'], errors='coerce')
support_tickets['date_closed'] = pd.to_datetime(support_tickets['date_closed'], errors='coerce')

# Calculate resolution time in days
support_tickets['resolution_days'] = (
    support_tickets['date_closed'] - support_tickets['date_opened']
).dt.days

# Keep only resolved tickets
resolved_tickets = support_tickets.dropna(subset=['resolution_days'])

# Group by issue category and calculate average resolution time
issue_resolution = resolved_tickets.groupby('issue_category').agg(
    avg_resolution_days=('resolution_days', 'mean')
).reset_index()

# Sort by longest resolution time
issue_resolution = issue_resolution.sort_values(
    by='avg_resolution_days', ascending=False
)

print(issue_resolution)
```

Q15: Which support agents have resolved the most critical tickets with high customer ratings (4)?

```python
critical_high_rating = support_tickets[
    (support_tickets['priority'] == 'Critical') &
    (support_tickets['customer_rating'] >= 4)
]

agent_performance = critical_high_rating.groupby('support_agent').agg(
    total_critical_tickets=('ticket_id', 'count')
).reset_index()

agent_performance = agent_performance.sort_values(
    by='total_critical_tickets', ascending=False
)

print(agent_performance)
```

Q25. What is the average loan duration for closed loans?

```python
closed_loans = loans[loans['loan_status'] == 'Closed'].copy()

closed_loans['start_date'] = pd.to_datetime(closed_loans['start_date'], errors='coerce')
closed_loans['end_date'] = pd.to_datetime(closed_loans['end_date'], errors='coerce')

closed_loans['loan_duration_days'] = (
    closed_loans['end_date'] - closed_loans['start_date']
).dt.days

avg_duration = closed_loans['loan_duration_days'].mean()
print("Average Loan Duration (days):", avg_duration)
```
