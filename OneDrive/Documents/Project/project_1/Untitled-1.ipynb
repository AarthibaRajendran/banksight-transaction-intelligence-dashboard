```python
import math
```

```python
def relu(x):
    return max(0,x)

def relu_derivative(x):
    return 1 if x > 0 else 0

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# input to Hidden (2 Neurons)
w1 = 0.5
w2 = 0.5

w3 = 0.4
w4 = 0.6

b1 = 0.2
b2 = 0.1

learning_rate = 0.1
```

```python
# Hidden to output
w5 = 0.7
w6 = 0.9

b3 = 0.3
```

```python
data = [
    (6,7,1), #Study, Sleep, Result
    (2,4,0),
    (5,5,1),
    (1,3,0)
]
```

```python
def forward_pass(study,sleep):
    z1 = study * w1 + sleep * w2 + b1
    z2 = study * w3 + sleep * w4 + b2

    h1 = relu(z1)
    h2 = relu(z2)

    z3 = h1 * w5 + h2 * w6
    output = sigmoid(z3)

    return z1 , z2 , h1 , h2 , z3 , output
```

```python
# Back Propagration
```

```python
for epoch in range(1000):

    total_loss = 0

    for study, sleep, actual in data:
        z1, z2, h1, h2, z3, pred = forward_pass(study, sleep)

        error = pred - actual
        total_loss += error ** 2

        # ---- Output layer ----
        d_pred = 2 * error
        d_sigmoid = pred * (1 - pred)

        dw5 = d_pred * d_sigmoid * h1
        dw6 = d_pred * d_sigmoid * h2
        db3 = d_pred * d_sigmoid

        # ---- Hidden layer ----
        dh1 = d_pred * d_sigmoid * w5
        dh2 = d_pred * d_sigmoid * w6

        dz1 = dh1 * relu_derivative(z1)
        dz2 = dh2 * relu_derivative(z2)

        dw1 = dz1 * study
        dw2 = dz1 * sleep
        db1 = dz1

        dw3 = dz2 * study
        dw4 = dz2 * sleep
        db2 = dz2

        # ---- Update weights ----
        w1 -= learning_rate * dw1
        w2 -= learning_rate * dw2

        w3 -= learning_rate * dw3
        w4 -= learning_rate * dw4

        w5 -= learning_rate * dw5
        w6 -= learning_rate * dw6

        b1 -= learning_rate * db1
        b2 -= learning_rate * db2
        b3 -= learning_rate * db3

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss {total_loss:.4f}")
```

```python
study = 6
sleep = 6

_ , _ , h1 , h2 , _ , result = forward_pass(study, sleep)
```
