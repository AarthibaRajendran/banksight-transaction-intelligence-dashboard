```python
#-------------------------------------------
# step 4 : create tables
#-------------------------------------------
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    customer_id VARCHAR(10) PRIMARY KEY,
    name VARCHAR(50),
    gender CHAR(1),
    age INT,
    city VARCHAR(50),
    account_type VARCHAR(20),
    join_date DATE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS accounts (
    customer_id VARCHAR(10),
    account_balance DECIMAL(12,2),
    last_updated DATETIME,
               
    FOREIGN KEY (customer_id)
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS transactions (
    txn_id VARCHAR(10) PRIMARY KEY,
    customer_id VARCHAR(10),
    txn_type VARCHAR(50),
    amount DECIMAL(12,2),
    txn_time DATETIME,
    status VARCHAR(50),
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS loans (
    loan_id INT PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id INT,
    branch VARCHAR(50),
    loan_type VARCHAR(50),
    loan_amount INT,
    interest_rate DECIMAL(12,2),
    loan_term_months INT,
    start_date DATE,
    end_date DATE,
    loan_status VARCHAR(50),
               
    FOREIGN KEY (customer_id)
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS credit_cards (
    card_id INT PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id INT,
    branch VARCHAR(50),
    card_number VARCHAR(20),
    card_type VARCHAR(50),
    card_network VARCHAR(50),
    credit_limit INT,
    current_balance DECIMAL(12,2),
    issued_date DATE,
    expiry_date DATE,
    status VARCHAR(50),
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers( customer_id)
    ON DELETE CASCADE
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS branches (
    branch_id INT PRIMARY KEY,
    branch_name VARCHAR(50),
    city VARCHAR(50),
    manager_name VARCHAR(50),
    total_employees int,
    branch_revenue DECIMAL(12,2),
    opening_date DATE,
    performance_rating int
)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS support_tickets (
    ticket_id VARCHAR(10) PRIMARY KEY,
    customer_id VARCHAR(10),
    account_id VARCHAR(10),
    loan_id VARCHAR(50) NULL,
    branch_name VARCHAR(50),
    issue_category VARCHAR(50),
    description  TEXT,
    date_opened DATE,
    date_closed DATE,
    priority VARCHAR(50),
    status VARCHAR(50),
    resolution_remarks TEXT,
    support_agent VARCHAR(50),
    channel VARCHAR(50),
    customer_rating int,
               
    FOREIGN KEY (customer_id) 
    REFERENCES customers(customer_id)
    ON DELETE CASCADE
)
""")
```

#### insert rows

```python
for _, row in customers.iterrows():
    cursor.execute("""
        INSERT INTO customers (customer_id, name, gender, age, city, account_type, join_date)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE
            name = VALUES(name),
            gender = VALUES(gender),
            age = VALUES(age),
            city = VALUES(city),
            account_type = VALUES(account_type),
            join_date = VALUES(join_date)
    """, (
        row['customer_id'], 
        row['name'], 
        row['gender'], 
        row['age'], 
        row['city'], 
        row['account_type'], 
        row['join_date']
    ))

conn.commit()
```
